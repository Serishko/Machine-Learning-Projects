```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\santo\PycharmProjects\MachineLearning\DataSets for ML\univariate_linear_regression\data.csv")
x = np.array(data.iloc[:,0])
y = np.array(data.iloc[:,1])
y
```

```python
alpha = 0.0007
iterations = 1000
theta0 = 0
theta1 = 0
m=len(y)
for i in range(iterations):
    ypred = theta0 + theta1*x
    theta0 = theta0 - alpha*(1/m)*sum(ypred-y)
    theta1 = theta1 - alpha*(1/m)*sum((ypred-y)*x)
    costfunc = (1/(2*m))*sum((ypred-y)**2)
    print("theta0:", theta0, "theta1:", theta1, "costfunc:", costfunc)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x,ypred, color='g')
ax2.scatter(x,y)
ax1.set_xlabel("x")
ax1.set_ylabel("ypred")
ax2.set_ylabel("y")
plt.show()
```

```python
# Using sklearn to compare the values:
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x.reshape(-1,1),y)
```

```python
list1=[model.predict(np.array(i).reshape(-1,1)) for i in x]
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, np.array(list1), color='g')
ax2.scatter(x, y)
ax1.set_xlabel("x")
ax1.set_ylabel("ypred")
ax2.set_ylabel("y")
plt.show()
```
