```python
import pandas as pd
import numpy as np
from math import exp, log
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\santo\PycharmProjects\MachineLearning\DataSets for ML\multivariate_logistic_regression\data.csv",names=['Studied','Slept','Passed'])
df
```

```python
# Categorizing data
studied = np.array(df['Studied'])
slept = np.array(df['Slept'])
y = np.array(df['Passed'])
```

```python
def predictions():
    ypredi=[]
    for i in range(len(y)):
        xvector = np.array([[1],[studied[i]],[slept[i]]])
        thetavector = np.array([[theta0],[theta1],[theta2]])
        z = np.matmul(np.transpose(thetavector),xvector)
        ypredi.append(1/(1+exp(-z[0])))
    return np.array(ypredi)

alpha = 0.1
m = len(y)
theta0 = 0
theta1 = 0
theta2 = 0
iterations = 10000
for i in range(iterations):
    ypred = predictions()
    theta0 = theta0 - (alpha/m)*sum(ypred-y)
    theta1 = theta1 - (alpha/m)*sum((ypred-y)*studied)
    theta2 = theta2 - (alpha/m)*sum((ypred-y)*slept)
    J = -(1/m)*sum(y*np.log(ypred)+(1-y)*np.log(1-ypred))
    print("theta0:{} theta1: {} theta2:{} J:{}".format(theta0,theta1,theta2,J))
```

```python
predicted_y = np.round(ypred)
```

```python
print("Final value of theta0, theta1 and theta2 are:",theta0,theta1,theta2)
```

```python
# Making df of passed and failed
df_passed = [row for index, row in df.iterrows() if row['Passed']==1]
df_failed = [row for index, row in df.iterrows() if row['Passed']==0]
df_passed = pd.DataFrame(df_passed)
df_failed = pd.DataFrame(df_failed)

# Plotting the merged graph
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(df_passed['Slept'], df_passed['Studied'],color='g',marker='o',label='Passed')
ax.scatter(df_failed['Slept'], df_failed['Studied'],color='b',marker='s',label='Failed')
plt.legend(loc='upper right')
ax.set_xlabel('Hours Slept')
ax.set_ylabel('Hours Studied')

# Drawing Decision Boundary
min_x1 = min(slept)-1
max_x1 = max(slept)+1
x1 = np.linspace(min_x1,max_x1)
x2 = (-theta0 - theta1*x1)/theta2
plt.plot(x1,x2)
plt.show()
```

```python
# using sklearn to predict
from sklearn import linear_model
model = linear_model.LogisticRegression()
model.fit(df[['Studied','Slept']],df['Passed'])
```

```python
print("values of theta1 and theta2 are:",model.coef_)
```

```python
print("value of theta0 is:",model.intercept_)
```

```python
# Comparing the scores between two models
our_score = np.sum(y == predicted_y)
sklearn_score = model.score(df[['Studied','Slept']],df['Passed'])*100
print("Our score is:{}\nSklearn score is:{}".format(sklearn_score, our_score))
```
